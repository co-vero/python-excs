```python
# EXC 1

def isMultiple(n: int) -> bool:
    if n % 3 == 0:
        return True
    elif n == 0:
        return False
    else:
        return isntMultiple(n-1)

def isntMultiple(n: int) -> bool:
    if n % 3 == 0:
        return False
    elif n == 0:
        return True
    else:
        return isMultiple(n-1)


print(isMultiple(12))
print(isntMultiple(12))
print(isMultiple(15))
```

```python
# EXC 2

def getSalaries():
    i = 0
    totalSalary = 0

    while True:
        name, salary = input("Enter name and salary, separated by a comma: ").split(",")

        if name == "fim":
            break
        else:
            name = name.strip()
            salary = float(salary.strip())

            totalSalary += salary
            i += 1
        
    if i == 0:
        return 0
    else:
        return f"{totalSalary / i :.2f}"

print(getSalaries())


```

```python
# EXC 3

def getHighest():
    i = 0
    highestSalary = 0

    while True:
        name, salary = input("Enter name and salary, separated by a comma: ").split(",")

        if name == "fim":
            break
        else:
            name = name.strip()
            salary = float(salary.strip())

            if salary > highestSalary:
                highestSalary = salary

            i += 1
        
    if i == 0:
        return 'Não tem'
    else:
        return f"{highestSalary:.2f}"

print(getHighest())
```

```python
# EXC 6

def crescimentoPopulacional():
    pa, pb, ta, tb = map(float, input("Enter population A, population B, rate A and rate B").split())

    pa = int(pa)
    pb = int(pb)

    if tb >= ta:
        return "A nunca alcançará B"

    anos = 0
    while pa <= pb:
        pa = int(pa * (1 + ta/100))
        pb = int(pb * (1 + tb/100))

        anos += 1

        if anos > 1000:
            return "Mais de um milênio"

    return "Vai alcançar em {} ano(s)".format(anos)


print(crescimentoPopulacional())
```

```python
# EXC 7

def fibonacci(n):
    a = 0
    b = 1
    if n < 1:
        return
    print(a, end=" ")
    for x in range(1, n):
        print(b, end=" ")
        c = a + b
        a = b
        b = c
    print()

fibonacci(2)
fibonacci(1)
fibonacci(10)
```

```python
# EXC 8

# TÁ ERRADO ESSE AQUI SOCORRO D:

# lê os dados de entrada
n, p = map(int, input().split())
nomes = input().split()

# cria a lista de alunos
alunos = []
for nome in nomes:
  alunos.append({"nome": nome, "amigos": []})

# adiciona as arestas
for i in range(p):
  nome1, nome2 = input().split()
  for aluno in alunos:
    if aluno["nome"] == nome1 and nome2 != nome1:
      aluno["amigos"].append(nome2)
    elif aluno["nome"] == nome2 and nome1 != nome2:
      aluno["amigos"].append(nome1)

# aplica a transitividade da amizade
for aluno in alunos:
  for amigo in aluno["amigos"]:
    for amigo_do_amigo in alunos:
      if amigo_do_amigo["nome"] == amigo:
        for novo_amigo in amigo_do_amigo["amigos"]:
          if novo_amigo not in aluno["amigos"] and novo_amigo != aluno["nome"]:
            aluno["amigos"].append(novo_amigo)

# remove os amigos duplicados
for aluno in alunos:
  aluno["amigos"] = list(set(aluno["amigos"]))

# imprime o resultado
for aluno in alunos:
  print(aluno["nome"], "possui", len(aluno["amigos"]))
```

```python
# EXC 17

def positivo(n):
    if n >= 0:
        return True
    else:
        return False
 
x = input()
acumulador = 0
count = 0
while x != "f":
    x = int(x)
    if not positivo(x):
        count += 1
        acumulador += x
    x = input()

if count > 0:
    media_negativo = acumulador/count
    print(media_negativo)
else:
    print("0.0") 
```
